2. In the above question, now scrape the following details of each product listed in first 3 pages of your
search results and save it in a data frame and csv. In case if any product has less than 3 pages in search
results then scrape all the products available under that product name. Details to be scraped are: "Brand
Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and
“Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“

```python
b_name=[]
p_name=[]
price=[]
ret_ex=[]
delivery=[]
avail=[]
url=[]

test=[]

#try :

for page in range(0,2):
    url_t =driver.find_elements(By.XPATH, '//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url_t:
        url.append(i.get_attribute('href'))
  
    nxt_button = driver.find_element(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')                                         

    nxt_button.click()                                                   

    time.sleep(5) 
     

for i in url:
    driver.get(i)
    time.sleep(2)

    try:
        b_name_t=driver.find_element(By.XPATH, '//tr[@class="a-spacing-small po-brand"]')
        b_name.append(b_name_t.text.split(" ")[1])
    except NoSuchElementException:
        b_name.append("----")
    try :
        p_name_t=driver.find_element(By.XPATH, '//span[@id="productTitle"]')
        p_name.append(p_name_t.text)
    except NoSuchElementException:
        p_name.append("----")
    try:
        price_t =driver.find_element(By.XPATH, '//span[@class="a-price a-text-price a-size-medium apexPriceToPay"]')
        price.append(price_t.text)
    except NoSuchElementException:
        price.append("----")
    try :
        ret_ex_t =driver.find_element(By.XPATH, '//div[@id="productSupportAndReturnPolicy-return-policy-message"]')
        ret_ex.append(ret_ex_t.text.split(" ")[1])
    except NoSuchElementException:
        ret_ex.append("----")
    try :
        delivery_t =driver.find_element(By.XPATH, '//div[@id="mir-layout-DELIVERY_BLOCK-slot-PRIMARY_DELIVERY_MESSAGE_LARGE"]')
        delivery.append(delivery_t.text)
    except NoSuchElementException:
        delivery.append("----")
    try : 
        avail_t =driver.find_element(By.XPATH, '//div[@id="availability"]')
        avail.append(avail_t.text)  
    except NoSuchElementException:
        avail.append("----")

driver.back()
```

```python
df1=pd.DataFrame({"Brand Name":b_name,"Name of the Product":p_name,"Price":price,"Expected Delivery":delivery,"Availability":avail,"Product URL":url})
df1
```

3. Write a python program to access the search bar and search button on images.google.com and scrape 10
images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’, ‘Guitar’, ‘Cakes’

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://images.google.com/")
time.sleep(5)

driver.maximize_window()
test=[]
dict1={'fruits':[], 'cars':[], 'Machine_Learning':[], 'Guitar':[], 'Cake':[]}

for i in ['fruits', 'cars', 'Machine_Learning', 'Guitar', 'Cake']:
    
    #scrolling down the page using pixel

    for _ in range(1000):
         driver.execute_script("window.scrollBy(0,1000000)")
      
    designation=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/input')
    designation.send_keys(i)
    
  

    search=driver.find_element(By.XPATH,'/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/button')
    search.click()
          
    time.sleep(5)
    
    
    img_t =driver.find_elements(By.XPATH, '//img[@class="rg_i Q4LuWd"]')
    for r in img_t:
        source=r.get_attribute('src') 
        if source is not None:
            if (source[0:4] == "http"):
                dict1[i].append(source)
               
                
        if len(dict1[i])>=11:
            break

    driver.back()
    time.sleep(1)  
    

df=pd.DataFrame(data=dict1)
df

for j in range(len(df.columns)):
    for i in range(10):
        print("Downloading {0} of {1} images".format(i,10))
        response=requests.get(df.iat[i,j])
        file=open(r"C:\Users\sumesh\OneDrive\Desktop\New folder\img"+str(j)+str(i)+".jpg","wb")
        file.write(response.content)
```

10. Write a python program to scrape a data for all available Hostels from https://www.hostelworld.com/ in
“London” location. You have to scrape hostel name, distance from city centre, ratings, total reviews,
overall reviews, privates from price, dorms from price, facilities and property description.

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.hostelworld.com/s?q=London,%20England&country=England&city=London&type=city&id=3&from=2022-12-23&to=2022-12-26&guests=2&page=1")
```

```python
name=[]
distance=[]
rating=[]
tr=[]
reviews=[]
price=[]
facilities=[]
desc=[]
x=0
y=0
z=0
url=[]

url_t=driver.find_elements(By.XPATH, '//h2[@class="title title-6"]//a')
for i in url_t:
    url.append(i.get_attribute('href'))
    
name_t=driver.find_elements(By.XPATH, '//h2[@class="title title-6"]')
for i in name_t:
    name.append(i.text)


distance_t=driver.find_elements(By.XPATH, '//span[@class="description"]')
for i in distance_t:
    distance.append(i.text)

rating_t=driver.find_elements(By.XPATH, '//div[@class="score gray big" or @class="score orange big"]')
for i in rating_t:
    rating.append(i.text)

tr_t=driver.find_elements(By.XPATH, '//div[@class="reviews"]')
for i in tr_t:
    tr.append(i.text)

price_t=driver.find_elements(By.XPATH, '//a[@class="prices"]')
for i in price_t[2:]:
    price.append(i.text.splitlines())
    
for i in url:
    driver.get(i)
    
    facilities_t=driver.find_elements(By.XPATH, '//ul[@class="facility-sections"]')
    for i in facilities_t:
        facilities.append(i.text.splitlines())

    desc_t=driver.find_elements(By.XPATH, '//div[@style="-webkit-line-clamp: 12;"]')
    for i in desc_t:
            desc.append(i.text.splitlines())
```

```python
hostel=pd.DataFrame({"hostel name":name,"Distance from City":distance,"Ratings":rating,"Total Reviews":tr,"Price of Privates and dorms":price,"facilities":facilities})
hostel
```
