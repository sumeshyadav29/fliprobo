1. Scrape the details of most viewed videos on YouTube from Wikipedia.
Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos
You need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]

r=driver.find_elements(By.TAG_NAME,"tr")
for i in r[2:32]:
    rank.append(i.text.split("\n")[0].split(".")[0])
    name.append(i.text.split("\n")[0].split(".")[1].split("[")[0])

c=4
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    artist.append(a[c].text)
    c=c+6

c=5
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    views.append(a[c].text)
    c=c+6

c=6
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    upload_date.append(a[c].text)
    c=c+6

details=pd.DataFrame({"Rank":rank,"Name":name,"Artist":artist,"Upload Date":upload_date,"Views":views})
details  
```

st ODI)

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.bcci.tv/")

search_button=driver.find_element(By.XPATH,"/html/body/nav/div[1]/div[2]/ul[1]/li[2]/a")
search_button.click()
```

```python

title=[]
series=[]
place=[]
date=[]
time=[]

t=driver.find_elements(By.XPATH,"//span[@class='matchOrderText ng-binding ng-scope']")
for i in t:
    title.append(i.text.split("-")[0])

s=driver.find_elements(By.XPATH,"//span[@class='ng-binding']")
for i in s:
    series.append(i.text)

p=driver.find_elements(By.XPATH,"//div[@class='fix-place ng-binding ng-scope']")
for i in p:
    place.append(i.text.split("-")[1])

d=driver.find_elements(By.XPATH,"//h5[@class='ng-binding']")
for i in d:
    date.append(i.text)

ti=driver.find_elements(By.XPATH,"//h5[@class='text-right ng-binding']")
for i in ti:
    time.append(i.text)  
    
schedule=pd.DataFrame(list(zip(title,series,place,date,time)),
                          columns=['Title','Series','Place','Date','Time'])
schedule 
```

```python
name=[]
artist=[]
last_week_rank=[]
peak_rank=[]
weeks=[]

a=driver.find_elements(By.XPATH,"//ul[@class='lrv-a-unstyle-list lrv-u-flex lrv-u-height-100p lrv-u-flex-direction-column@mobile-max']")

artist.append(a[0].text.split("\n")[1])
last_week_rank.append(a[0].text.split("\n")[2])
peak_rank.append(a[0].text.split("\n")[3])
weeks.append(a[0].text.split("\n")[4])


a=driver.find_elements(By.TAG_NAME,"h3")
for i in a[1:]:
    name.append(i.text)
while ("" in name):
    name.remove("")    
 


a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-no-trucate a-font-primary-s lrv-u-font-size-14@mobile-max u-line-height-normal@mobile-max u-letter-spacing-0021 lrv-u-display-block a-truncate-ellipsis-2line u-max-width-330 u-max-width-230@tablet-only']")
for i in a:
    artist.append(i.text)
 


j=0
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    last_week_rank.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break



j=1
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    peak_rank.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break



j=2
a=driver.find_elements(By.XPATH,"//span[@class='c-label  a-font-primary-m lrv-u-padding-tb-050@mobile-max']")
for i in range(len(a)):
    weeks.append(a[j].text)
    j=j+3
    if (len(a)-1)<=j:
        break
while "" in last_week_rank: 
    last_week_rank.remove("")
while "" in peak_rank: 
    peak_rank.remove("")
while "" in weeks: 
    weeks.remove("")
    
song_details=pd.DataFrame(list(zip(name[:100],artist,last_week_rank,peak_rank,weeks)),
                          columns=['Song Name','Artist','Last Week Rank','Peak rank','Weeks on board'])
song_details  
```

Q9)Scrape the details most watched tv series of all time from imdb.com.Url = https://www.imdb.com/list/ls095964455/ You have to find the following details: A) Name B) Year span C) Genre D) Run time E) Ratings F) Votes

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://www.imdb.com/list/ls095964455/")
```

```python
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
votes=[]

a=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']")
for i in a:
    name.append(i.text.split(".")[1].split("(")[0])
    

a=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
for i in a:
    year_span.append(i.text.split("(")[1].split(")")[0])

a=driver.find_elements(By.XPATH,"//span[@class='genre']")
for i in a:
    genre.append(i.text)

a=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in a:
    run_time.append(i.text)

a=driver.find_elements(By.XPATH,"//span[@class='ipl-rating-star__rating']")
for i in a:
    if i.text=="\n" or i.text=="Rate" or i.text=="":
        continue
    ratings.append(i.text)  
    

a=driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']")

for i in range(2,len(a),3):
    votes.append(a[i].text.split(":")[1])
    
show_details=pd.DataFrame(list(zip(name,year_span,genre,run_time,ratings,votes)),
                          columns=['Show Name','Year of Span','Genre','Run Time','Rating','Votes'])
show_details 
```

Q10) Details of Datasets from UCI machine learning repositories.Url = https://archive.ics.uci.edu/ You have to find the following details: A) Dataset name B) Data type C) Task D) Attribute type E) No of instances F) No of attribute G) Year

```python
driver=webdriver.Chrome(r"chromedriver.exe")
driver.get("https://archive.ics.uci.edu/")

search_button=driver.find_element(By.XPATH,"/html/body/table[2]/tbody/tr/td/span/b/a")
search_button.click()
```

```python
name=[]
data_type=[]
task=[]
att_type=[]
instances=[]
attribute=[]
year=[]


a=driver.find_elements(By.XPATH,"//p[@class='normal']")
for i in range(8,len(a[8:]),7):
    name.append(a[i].text)

for i in range(9,len(a[9:]),7):
    data_type.append(a[i].text)
    

for i in range(10,len(a[10:]),7):
    task.append(a[i].text)
    

for i in range(11,len(a[11:]),7):
    
    att_type.append(a[i].text)

for i in range(12,len(a[12:]),7):
    
    instances.append(a[i].text)
    
for i in range(13,len(a[13:]),7):
    
    attribute.append(a[i].text)

for i in range(14,len(a[14:]),7):
    
    year.append(a[i].text)
    
show_details=pd.DataFrame(list(zip(name,data_type,task,att_type,instances,attribute,year)),
                          columns=['Dataset Name','Data Type','Task','Attribute Type','Instances','Attribute','Year'])
show_details  
```
