2.Write python program to display IMDB's Top rated 100 movies data(name, rating, year of release) and make dataframe

```python
page1= requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
imdb1=BeautifulSoup(page1.content,"html.parser")
```

```python
alldata={"name":[],"year":[],"rating":[]}
for n in imdb1.find_all("td","titleColumn"):
    alldata["name"].append(n.get_text().replace("\n","").split("(")[0])
    alldata["year"].append(n.get_text().replace("\n","").split("(")[1].replace(")",""))
    
for i in imdb1.find_all("td","ratingColumn imdbRating"):
    alldata["rating"].append(i.get_text().replace("\n",""))
    
```

```python
df = pd.DataFrame(data=alldata)
imdb_top_100=df.head(100)
```

```python
imdb_top_100
```

5.Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:

a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
page5a= requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
icc_team=BeautifulSoup(page5a.content,"html.parser")
```

```python
teams=[]

for i in icc_team.find_all("span","u-hide-phablet"):
    teams.append(i.get_text())
    
matches=[]
matches.append((icc_team.find("td", "rankings-block__banner--matches")).text)
x=1
for i in icc_team.find_all("td","table-body__cell u-center-text"):
    if x%2==1:
        matches.append(i.text)
    x=x+1

points=[]
points.append((icc_team.find("td", "rankings-block__banner--points")).text)
x=1
for i in icc_team.find_all("td","table-body__cell u-center-text"):
    if x%2==0:
        points.append(i.text)
    x=x+1
    
rating=[]
rating.append((icc_team.find("td", "rankings-block__banner--rating u-text-right")).text.split("\n")[1].strip())
for i in icc_team.find_all("td","table-body__cell u-text-right rating"):
    rating.append(i.get_text())
```

```python
dfteam=pd.DataFrame({"Team" : teams, "Matches":matches, "Points": points, "Rating": rating})
teams_10=dfteam.head(10)
```

```python
teams_10
```

8) Write a python program to scrape the details of most downloaded articles from AI in last 90 days.
https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles
Scrape below mentioned details :
i) Paper Title
ii) Authors
iii) Published Date
iv) Paper URL 

```python
page8= requests.get("https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles")
ai=BeautifulSoup(page8.content,"html.parser")
```

```python
dict_ai={"Title":[],"Author":[],"Published Date":[], "Paper URL":[]}

for n in ai.find_all("h2",class_="sc-1qrq3sd-1 gRGSUS sc-1nmom32-0 sc-1nmom32-1 btcbYu goSKRg"):
    dict_ai["Title"].append(n.text)
    
for i in ai.find_all("span",class_="sc-1w3fpd7-0 dnCnAO"):
    dict_ai["Author"].append(i.get_text())
    
for i in ai.find_all("span",class_="sc-1thf9ly-2 dvggWt"):
    dict_ai["Published Date"].append(i.get_text())
    
for i in ai.find_all("a", "sc-5smygv-0 fIXTHm"):
    dict_ai["Paper URL"].append(i.get("href"))
```

```python
df_ui= pd.DataFrame(data=dict_ai)
df_ui
```

Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
page9= requests.get("https://www.dineout.co.in/delhi-restaurants/welcome-back")
dine=BeautifulSoup(page9.content,"html.parser") 
```

```python
dict_dine={"Restaurant name":[],"Cuisine":[],"Location":[], "Ratings":[], "Image URL":[]}

for n in dine.find_all("a",class_="restnt-name ellipsis"):
    dict_dine["Restaurant name"].append(n.text)
    
for i in dine.find_all("span",class_="double-line-ellipsis"):
    dict_dine["Cuisine"].append(i.get_text().split("|")[1])
    
for i in dine.find_all("div",class_="restnt-loc ellipsis"):
    dict_dine["Location"].append(i.get_text())
    
for i in dine.find_all("div", "restnt-rating rating-4"):
    dict_dine["Ratings"].append(i.get_text())
    
for i in dine.find_all("img", "no-img"):
    dict_dine["Image URL"].append(i.get("data-src"))
```

```python
dine_df=pd.DataFrame(data=dict_dine)
dine_df
```

10) Write a python program to scrape the details of top publications from Google Scholar from
https://scholar.google.com/citations?view_op=top_venues&hl=en
i) Rank
ii) Publication
iii) h5-index
 iv) h5-median

```python
page10= requests.get("https://scholar.google.com/citations?view_op=top_venues&hl=en")
pub=BeautifulSoup(page10.content,"html.parser")
```

```python
dict_pub={"Rank":[],"Publication":[],"h5-index":[], "h5-median":[]}

for n in pub.find_all("td",class_="gsc_mvt_p"):
    dict_pub["Rank"].append(n.text)
    
for i in pub.find_all("td",class_="gsc_mvt_t"):
    dict_pub["Publication"].append(i.get_text())
    
for i in pub.find_all("a",class_="gs_ibl gsc_mp_anchor"):
    dict_pub["h5-index"].append(i.get_text())
    
for i in pub.find_all("span", "gs_ibl gsc_mp_anchor"):
    dict_pub["h5-median"].append(i.get_text())
```

```python
pub_df=pd.DataFrame(data=dict_pub)
pub_df
```
